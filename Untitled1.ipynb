```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import warnings
```

```python
response = requests.get('https://books.toscrape.com/')
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
books = soup.select('.product_pod')

matrix = []

for book in books:
  title = book.find('a').find('img').get('alt')
  price = float(book.select_one('.price_color').text.strip()[2:])
  matrix.append([title, price])
  # print(books.find('a').find('img').get('alt'))
  # print(books.select_one('.price_color').text.strip()[1:])

df = pd.DataFrame(matrix, columns=["Title", "Price"])
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.hist(df['Price'], bins=30)
plt.show()
```
